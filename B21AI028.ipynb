```python
# Write a function to handle the missing values in the dataset (e.g., any NA, NaN values).
import pandas as pd

def handle_missing_values(df):
    # Handle missing values by replacing with the mean of the column
    df = df.fillna(df.mean())
    
    # replacing missing values by the mode of the column
    for col in df.columns:
        df[col].fillna(df[col].mode()[0], inplace=True)
    
    return df
handle_missing_values(df)
  
```

```python
#Write a function to reduce noise (any error in the feature) in individual attributes 

import numpy as np

def reduce_noise(df, attribute, std_factor=3):
    mean = df[attribute].mean()
    std = df[attribute].std()
    threshold = std_factor * std
    
    # Remove values that are more than 3 standard deviations from the mean
    df[attribute] = df[attribute][np.abs(df[attribute]-mean) <= threshold] 
    # Replace remaining missing values with the mean
    df[attribute].fillna(mean, inplace=True)
    
    return df
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LogisticRegression as LR
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.metrics import confusion_matrix as cm
from sklearn.metrics import precision_score as ps
from sklearn.metrics import recall_score as rs
from sklearn.metrics import f1_score as f1s
from sklearn.metrics import accuracy_score as acc

```

```python
data = pd.read_csv("/content/drive/MyDrive/PRML LABs/lab 1/data.csv")
data
```

```python
condition_M = data.diagnosis == "M"
condition_B = data.diagnosis == "B"

data.loc[condition_M,"diagnosis"]=0
data.loc[condition_B,"diagnosis"]=1

data
```

```python
Y = data.diagnosis.to_numpy().astype('int')                                     # Labels

X_data = data.drop(columns=["id","diagnosis","Unnamed: 32"])
X = X_data.to_numpy()
  
```

```python
user_prompt = 0.3
user_enable = False

x_train,x_test,y_train,y_test = tts(X,Y,test_size=user_prompt,shuffle=user_enable)
```

```python
# Note :- Don't worry about the code snippet here, it is just to produce the predictions for the test data portion of each classifier

logistic_model = LR()
logistic_model.fit(x_train,y_train)
logistic_pred = logistic_model.predict(x_test)

decision_model = DTC()
decision_model.fit(x_train,y_train)
decision_pred = decision_model.predict(x_test)
```

```python
#avg_accuracy
inbuilt_acc_logistic = acc(y_test,logistic_pred)
inbuilt_acc_decision = acc(y_test,decision_pred)

print("Accuracy for Logistic Regression-based Predictions =>",str(inbuilt_acc_logistic*100)+"%")
print("Accuracy for Decision Tree-based Predictions =>",str(inbuilt_acc_decision*100)+"%")

def avg_accuracy(y_true, y_pred):
    true_positive = 0
    true_negative = 0
    false_positive =0
    false_negative = 0

    for i in range(len(y_true)):
        if y_true[i] == 1 and y_pred[i] == 1:
          true_positive += 1
        elif y_true[i] == 0 and y_pred[i] == 0:
          true_negative += 1
        elif y_true[i] == 0 and y_pred[i] == 1:
          false_positive += 1
        else:
          false_negative += 1

    avg_accuracy = (true_positive+true_negative) / (true_positive +true_negative+ false_negative+false_positive)
    return avg_accuracy
b = avg_accuracy(y_test,decision_pred)
a = avg_accuracy(y_test,logistic_pred)
print("Accuracy for Logistic Regression-based Predictions =>",str(a*100)+"%")
print("Accuracy for Decision Tree-based Predictions =>",str(b*100)+"%")
```

```python
#Precision
inbuilt_ps_logistic = ps(y_test,logistic_pred)
inbuilt_ps_decision = ps(y_test,decision_pred)

print("Precision for Logistic Regression-based Predictions =>",str(inbuilt_ps_logistic*100)+"%")
print("Precision for Decision Tree-based Predictions =>",str(inbuilt_ps_decision*100)+"%")

def precision(y_true, y_pred):
    true_positive = 0
    true_negative = 0
    false_positive =0
    false_negative = 0

    for i in range(len(y_true)):
        if y_true[i] == 1 and y_pred[i] == 1:
          true_positive += 1
        elif y_true[i] == 0 and y_pred[i] == 0:
          true_negative += 1
        elif y_true[i] == 0 and y_pred[i] == 1:
          false_positive += 1
        else:
          false_negative += 1

    precision = (true_positive) / (true_positive +false_positive)
    return precision
b = precision(y_test,decision_pred)
a = precision(y_test,logistic_pred)
print("Precision for Logistic Regression-based Predictions =>",str(a*100)+"%")
print("Precision for Decision Tree-based Predictions =>",str(b*100)+"%")
```

```python
#Recall
inbuilt_rs_logistic = rs(y_test,logistic_pred)
inbuilt_rs_decision = rs(y_test,decision_pred)
print("Recall for Logistic Regression-based Predictions =>",str(inbuilt_rs_logistic*100)+"%")
print("Recall for Decision Tree-based Predictions =>",str(inbuilt_rs_decision*100)+"%")

def recall(y_true, y_pred):
    true_positive = 0
    false_negative = 0
    for i in range(len(y_true)):
        if y_true[i] == 1 and y_pred[i] == 1:
            true_positive += 1
        elif y_true[i] == 1 and y_pred[i] == 0:
            false_negative += 1
    recall = true_positive / (true_positive + false_negative)
    return recall
b = recall(y_test,decision_pred)
a  = recall(y_test,logistic_pred)
print("Recall for Logistic Regression-based Predictions =>",str(a*100)+"%")
print("Recall for Decision Tree-based Predictions =>",str(b*100)+"%")
```
