```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

```python
bcd = load_breast_cancer()
bcd
```

```python
df = pd.DataFrame(bcd.data, columns = bcd.feature_names)
df.head()
```

```python
df['label'] = bcd.target
```

```python
x = df.drop(columns='label',axis =1) # droup column axis = 1 o.w. axis =0
y = df['label']
```

```python
x
```

```python
# y
```

```python
x_train ,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=23) # random_state reporduceable result
```

```python
print(x.shape,x_test.shape,x_train.shape)
```

```python
#standarize the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)

x_test_std = scaler.transform(x_test)
```

## Build the Neural Network
>> importing libraries - tensorflow , keras

```python
import tensorflow as tf
tf.random.set_seed(3) #for same accuracy score
from tensorflow import keras
```

```python
# setting up the layer of nn

model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)), # for input layer // convert data into the single dimention array shape
    keras.layers.Dense(20,activation='relu') ,# for output layer
    keras.layers.Dense(2,activation = 'sigmoid') # for output layer 
    ])
```

```python
#compiling the nn optimizer 
model.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy',metrics= ['accuracy'])
```

```python
# training the nn 
history = model.fit(x_train_std,y_train,validation_split=0.1,epochs=10) # epochs === how many times your data go though the model
```

Visualization of loss and accuracy

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')

plt.legend(['training data','validation data'],loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.xlabel('epochs')
plt.ylabel('loss')

plt.legend(['training data','validation data'],loc = 'lower right')
```

Accuracy of model on test data

```python
loss , accuracy = model.evaluate(x_test_std,y_test)
print(loss)
print(accuracy)
```
